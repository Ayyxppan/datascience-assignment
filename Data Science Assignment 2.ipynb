```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="ball",data=data)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="ball",data=data)
```

```python
sns.boxplot(x="over",y="ball",data=data)
plt.title("balls per over")
plt.xlabel("over")
plt.ylabel("ball")
```

```python
sns.lineplot(x="over",y="ball",data=data)
plt.title("balls per over")
plt.xlabel("over")
plt.ylabel("ball")
```

```python
sns.scatterplot(x="over",y="ball",data=data, color="pink")
plt.title("balls per over")
plt.xlabel("over")
plt.ylabel("ball")
```
